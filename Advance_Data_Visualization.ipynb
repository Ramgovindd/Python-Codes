```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
iris=snr.load_dataset('iris')
```

```python
df=iris.drop(['species'], axis=1)
```

```python
df.head()
```

Scatter Plot

```python
df.plot.scatter(x='sepal_length', y='petal_length')
```

```python
df.plot.scatter(x='petal_length', y='sepal_width')
```

```python
df.plot.scatter(x='sepal_length', y='petal_length', c='sepal_width')
```

```python
df.plot.scatter(x='petal_length', y='sepal_width', c='sepal_length',s=190) #s is the size of datapoint
```

Swarm Plot

```python
df.plot.hexbin(x='petal_length', y='sepal_width', gridsize=25, C='sepal_length')
```
