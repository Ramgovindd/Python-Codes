Visulaization of Data in the following format:

1. Bar Graph

2. Histogram

3. Pie Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #this is used for data visualization
import seaborn as snr #this is used for data visualization
```

```python
#loading the IRIS data 
iris=snr.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.hist(color='b',alpha=0.7,figsize=(8,8)) #alpha controls the intensity
```

```python
iris.plot(kind='bar')
```

```python
iris.plot(kind='hist')
```

```python
iris['sepal_length'].plot(kind='hist', bins=50) #width of each vertical histogram is maintained by bins
```

```python
iris['petal_length'].plot(kind='hist', bins=50) #width of each vertical histogram is maintained by bins
```

```python
iris['sepal_width'].plot(kind='hist', bins=50) #width of each vertical histogram is maintained by bins
```

```python
iris['petal_width'].plot(kind='hist', bins=50) #width of each vertical histogram is maintained by bins
```

```python
df=iris.drop(['species'],axis=1) #axis=1 means all the values inside the column will vanish
```

```python
df.head()
```

```python
df.plot.pie(subplots=True,figsize=(20,20),fontsize=26)
```

```python
new_df=df.head(5) #first 5 rows of my df data will be added t new_df dataset
```
